```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import seaborn as sns
import matplotlib.pyplot as plt

QUERY = """SELECT ConvertedComp FROM master"""
ConvertedComp_df = pd.read_sql_query(QUERY,conn)
plt.hist(ConvertedComp_df, bins=30, edgecolor='black')

# Set labels and title
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.title('Histogram of Converted Compensation')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """SELECT Age FROM master"""
Age_df = pd.read_sql_query(QUERY,conn)
sns.boxplot(Age_df)
plt.title('Box Plot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """SELECT Age,WorkWeekHrs  FROM master"""
Age_WorkWeekHrs_df = pd.read_sql_query(QUERY,conn)
sns.scatterplot(data=Age_WorkWeekHrs_df,x='Age',y='WorkWeekHrs')
plt.title('Age-WorkWeekHrs scatterplot')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """SELECT Age,WorkWeekHrs,CodeRevHrs  FROM master"""
Ag_Work_Code_df = pd.read_sql_query(QUERY,conn)
sns.scatterplot(data=Ag_Work_Code_df,x='CodeRevHrs',y='WorkWeekHrs',size ='Age',sizes=(50, 500), alpha=0.7)
plt.title('WorkWeekHrs-CodeRevHrs-Age bubbleplot')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# print all tables 

QUERY = """
SELECT *
FROM sqlite_master

"""
df = pd.read_sql_query(QUERY,conn)
df
```

```python
# your code goes here
QUERY = """
SELECT *, COUNT(*) AS count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5

"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(8, 6))
plt.pie(df['count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140, shadow=True)


plt.title('Top 5 Databases Respondents Wish to Learn Next Year')


plt.axis('equal')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """SELECT Age,WorkWeekHrs,CodeRevHrs  FROM master WHERE Age BETWEEN 30 AND 35"""
Ag_Work_Code_df = pd.read_sql_query(QUERY,conn)
median_data = Ag_Work_Code_df.groupby('Age').median()
plt.figure(figsize=(10, 6))
plt.bar(median_data.index,median_data['WorkWeekHrs'],label='Median WorkWeekHrs')
plt.bar(median_data.index,median_data['CodeRevHrs'],bottom=median_data['WorkWeekHrs'], label='Median CodeRevHrs')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.legend()


plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """SELECT Age,ConvertedComp  FROM master WHERE Age BETWEEN 45 AND 60"""
Ag_Conv_df = pd.read_sql_query(QUERY,conn)
median_data1 =Ag_Conv_df.groupby('Age').median()

plt.figure(figsize=(10, 6))
plt.plot(median_data1.index, median_data1['ConvertedComp'], marker='o', linestyle='-')


plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')


plt.grid(True)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master
"""
main_branch_df = pd.read_sql_query(QUERY, conn)

main_branch_counts = main_branch_df['MainBranch'].value_counts()


plt.figure(figsize=(10, 6))
plt.barh(main_branch_counts.index, main_branch_counts)


plt.title('Number of Respondents by MainBranch')
plt.xlabel('Number of Respondents')
plt.ylabel('MainBranch')


plt.show()






```
